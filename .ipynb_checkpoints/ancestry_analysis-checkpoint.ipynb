```python
%%R

library(ggthemes)
library(ggplot2)
library(reshape2)
library(gridExtra)
library(ALDEx2)
library(RColorBrewer)

ascols = function(facs, pallette){
    ffacs = as.factor(as.character(facs))
    n = length(unique(facs))
    cols = pallette(n)[ffacs]
    return(cols)
}
```

```python

def from_percent(val):
    if isinstance(val, (int, float)):
        return val
    if '%'  in val:
        val = val.split('%')[0]
    try:
        return int(val)
    except ValueError:
        return val

raw_ancestry = pd.read_csv('/home/dcdanko/Dropbox/resources_and_shared/metasub_data_packets/metasub_data_packet_v1_0/other/raw_ancestry_results.csv').apply(lambda x: x.apply(from_percent))
raw_ancestry.head()
```

```python
ancestry_cols = [
 'Africa1',
 'Africa2',
 'Africa3',
 'EastAsian1',
 'EastAsian2',
 'European1',
 'European2',
 'Iberia',
 'SouthAmerica1',
 'SouthAsia',
 'WestEurope',
 'Africans',
]

def center(col):
    try:
        return col - col.mean()
    except ValueError:
        return col
    
cities = raw_ancestry['City']
original = raw_ancestry[ancestry_cols]
remapped = raw_ancestry[ancestry_cols].apply(lambda s: (s - s.mean()) > 0).apply(lambda s: s.apply(lambda x: )) # .apply(lambda s: (s - s.mean()) / s.std())
print(remapped.shape)
remapped.head()
```
