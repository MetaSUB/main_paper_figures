```python
%load_ext rpy2.ipython

import pandas as pd
from metasub_utils.packet_parse import MetaSUBTableFactory
from capalyzer.packet_parser.experimental import umap
import warnings
from os.path import isfile

warnings.filterwarnings('ignore')

tabler = MetaSUBTableFactory.core_factory()
```

```python
%%R

library(ggthemes)
library(ggplot2)
library(reshape2)
library(gridExtra)
library(ALDEx2)
library(RColorBrewer)
```

```python
taxa = tabler.taxonomy(normalize=True)
taxa = taxa[['Staphylococcus aureus', 'Staphylococcus lugdunensis']]
taxa.head()
```

```python
%%R -i taxa

ggplot(taxa, aes(x=Staphylococcus.aureus, y=Staphylococcus.lugdunensis)) +
    geom_point() +
    scale_y_log10() +
    scale_x_log10() +
    theme(
        text = element_text(size=20),
    )
```

```python

def b_given_a(tbl, threshb=0, thresha=0):
    a = tbl['Staphylococcus aureus'] > thresha
    b = tbl['Staphylococcus lugdunensis'] > threshb
    return (b & a).mean()

def b_given_nota(tbl, threshb=0, thresha=0):
    a = tbl['Staphylococcus aureus'] > thresha
    b = tbl['Staphylococcus lugdunensis'] > threshb
    return (b & ~a).mean()
    
    
def one_iter(taxa, threshb=0, thresha=0):
    taxa = taxa.sample(1000)
    return b_given_nota(taxa, threshb=threshb, thresha=thresha), b_given_a(taxa, threshb=threshb, thresha=thresha), thresha, threshb

pts = []
for ia in [-1, -0.5, 0, 0.5, 1, 2, 3]:
    ta = taxa['Staphylococcus aureus'].mean() - ia * taxa['Staphylococcus aureus'].std()
    ta = max(0, ta)
    ta = min(1, ta)
    for ib in [-1, -0.5, 0, 0.5, 1, 2, 3]:
        tb = taxa['Staphylococcus lugdunensis'].mean() - ib * taxa['Staphylococcus lugdunensis'].std()
        tb = max(0, tb)
        tb = min(1, tb)
        pts += [list(one_iter(taxa, threshb=tb, thresha=ta)) + [ia, ib] for _ in range(1000)]
        
pts = pd.DataFrame(pts)
pts.columns = ['X', 'Y', 'threshA', 'threshB', 'sdA', 'sdB']
pts.head()
```

```python
%%R -i pts -w 1080 -h 540

pts$E = log2(pts$Y / pts$X)
ggplot(pts, aes(x=sdA, y=sdB, color=E)) +
    geom_jitter() +
    scale_color_distiller(palette='Spectral', na.value='white') +
    theme_dark() +
    theme(
        text = element_text(size=20),
    )
```

```python
from scipy.spatial.distance import pdist, squareform
from itertools import combinations

FNAME = 

def eco_val(c1, c2):
    a = c1 > c1.mean()
    b = c2 > c2.mean()
    return (b & a).mean() / (b & ~a).mean()    
    
    
def one_iter(taxa):
    taxa = taxa.sample(1000)
    tbl = {}
    for c1_name in taxa.columns:
        tbl[c1_name] = {}
        for c2_name in taxa.columns:
            tbl[c1_name][c2_name] = eco_val(taxa[c1_name], taxa[c2_name])
    return pd.DataFrame.from_dict(tbl, orient='index')
    

def eco_dists(taxa, niter):
    D = one_iter(taxa)
    for i in range(niter - 1):
        print(f'iteration {i}')
        D += one_iter(taxa)
    D /= niter
    return D
    
taxa = tabler.taxonomy(normalize=True)
taxa = taxa[taxa.columns[(taxa > 0).mean() > 0.70]]

taxa_dists = eco_dists(taxa, 10)
```

```python
import numpy as np

taxa_dists = taxa_dists.replace([np.inf, -np.inf], np.nan).fillna(1).applymap(np.log2)

taxa_dists
```

```python
import seaborn as sns

sns.clustermap(
    taxa_dists, center=0.35, cmap="RdBu_r",
    figsize=(40, 40)
)
```
