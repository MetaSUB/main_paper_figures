```python
%load_ext rpy2.ipython

import pandas as pd
from metasub_utils.packet_parse import MetaSUBTableFactory
from capalyzer.packet_parser.experimental import umap
from capalyzer.packet_parser import DataTableFactory
import warnings
from os.path import isfile

warnings.filterwarnings('ignore')

metadata = pd.read_csv('~/Dropbox/resources_and_shared/release_packet_jun12_2019/metadata/complete_metadata.csv', index_col=0)
#metadata = metadata.loc[metadata['control_type_coarse'].isna()]
metadata = metadata.query('city != "montevideo"')
metadata['city'] = metadata['city'].map(str)

core_tabler = MetaSUBTableFactory('~/Dropbox/resources_and_shared/release_packet_jun12_2019', metadata_tbl=metadata)

sou_metadata = pd.read_excel('/home/dcdanko/Downloads/sou_abrf_packet_jan_14_2020/metadata/abrf-clean-metadata.xlsx', index_col=1)
names ={el for el in sou_metadata.index if 'P1' in str(el)}
sou_metadata = sou_metadata.loc[names]
def cleanit(el):
    #el = el.split('_')[1]
    if 'P' in el:
        el = el.split('P')[1]
    if el[0] not in 'ABCDEFGH':
        el = el[1:]
    if len(el) == 2:
        el = el[0] + '0' + el[1]
    return el


sou_metadata.index = sou_metadata.index.map(cleanit)
names = set()
for p in 'ABCDEFGH':
    for s in range(1, 13):
        el = p + str(s)
        if len(el) == 2:
            el = el[0] + '0' + el[1]
        names.add(el)

#sou_metadata = sou_metadata.loc[names]
sou_tabler = DataTableFactory('/home/dcdanko/Downloads/sou_abrf_packet_jan_14_2020', metadata_tbl=sou_metadata)

from sklearn.decomposition import PCA
from sklearn.ensemble import RandomForestClassifier
from sklearn import svm 
from sklearn.svm import LinearSVC
from sklearn import svm 
from multiprocessing import Pool

import numpy as np

from effet_oppos.preprocessing import (
    parse_raw_data,
    parse_feature,
    normalize_data,
)
from effet_oppos.classification import (
    split_data,
    get_classifier,
    k_fold_crossvalid,
    train_model,
    predict_with_model,
    multi_predict_with_model,
    predict_top_classes,
    feature_importance,
)

from sklearn.metrics import (
    precision_score,
    recall_score,
    confusion_matrix,
    classification_report,
    accuracy_score,
)
```

```python

def train_eval(cities, taxa):
    feature, name_map = pd.factorize(cities)
    normalized = normalize_data(taxa, method='binary', threshold='0.0001')
    train_taxa, test_taxa, train_feature, test_feature = split_data(
            normalized, feature, test_size=0.22
    )
    model = train_model(
            train_taxa, train_feature, method='random_forest',
            n_estimators=40, n_neighbours=21, n_components=10, seed=None
    )
    predictions = predict_with_model(model, test_taxa).round()
    return {
        'accuracy': accuracy_score(test_feature, predictions),
        'precision': precision_score(test_feature, predictions, average="macro"),
        'recall': recall_score(test_feature, predictions, average="macro"),
        'train_size': train_taxa.shape[0],
        'test_size': test_taxa.shape[0],
    }

sou_taxa = sou_tabler.taxonomy()
sou_cities = sou_metadata.loc[sou_taxa.index, 'OriginCity_Cleaned']
sou_cities = sou_cities.dropna()
sou_cities = sou_cities[~sou_cities.index.duplicated(keep='first')]
sou_cities = sou_cities.map(lambda el: 'SOU ' + str(el))
foo = sou_cities

foo = foo.map(lambda el: el.replace(' ', '_').lower())
foo = foo.map(lambda el: '_'.join(el.split('_')[1:]).split(',')[0])
foo = foo.map(lambda el: 'new_york_city' if el == 'new_york' else el)
sou_cities = foo

cities = set(sou_cities) & set(msub_cities)
cities = ['new_york_city', 'brisbane']

sou_cities = sou_cities[sou_cities.isin(cities)]

sou_taxa = sou_taxa.loc[sou_cities.index]

results = []
for _ in range(100):
    results.append(train_eval(sou_cities, sou_taxa))
    
results = pd.DataFrame(results)
results.head()
```

```python
def one():
    cities = pd.concat([sou_cities.sample(frac=1), msub_cities.sample(frac=1)])
    feature, name_map = pd.factorize(cities)
    taxa = pd.concat([sou_taxa, msub_taxa]).fillna(0)
    taxa = taxa.loc[cities.index]


    N, M = 11, 1000
    test_cities = cities.iloc[:N].iloc[:5]
    test_feature = feature[:N][:5]
    test_taxa = taxa.iloc[:N].iloc[:5]
    train_cities = cities.iloc[N:M]
    train_feature = feature[N:M]
    train_taxa = taxa.iloc[N:M]
    
    model = train_model(
            train_taxa, train_feature, method='random_forest',
            n_estimators=40, n_neighbours=21, n_components=10, seed=None
    )
    predictions = predict_with_model(model, test_taxa).round()
    result = {
        'accuracy': accuracy_score(test_feature, predictions),
        'precision': precision_score(test_feature, predictions, average="macro"),
        'recall': recall_score(test_feature, predictions, average="macro"),
        'train_size': train_taxa.shape[0],
        'test_size': test_taxa.shape[0],
    }
    return result

msub_results = []
for _ in range(100):
    result = one()
    msub_results.append(result)
msub_results = pd.DataFrame(msub_results)

msub_results.head()
```
